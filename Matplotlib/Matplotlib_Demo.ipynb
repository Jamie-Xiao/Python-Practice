## 1) Bar chart
Bar plot the most popular programming languages

```python
import csv
from matplotlib import pyplot as plt 
from collections import Counter
```

```python
cnter = Counter()
languages = []
people = []

#read and count the data
with open('data/data.csv',encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    for item in csv_reader:
        cnter.update(item['LanguagesWorkedWith'].split(';'))

#prepare for x and y axis
for tup in cnter.most_common(10):
    languages.append(tup[0])
    people.append(tup[1])

print(languages)    
print(people)    
```

```python
#prepare the plot parameter
plt.title('The Most Popular Languages')
plt.xlabel('Programing Language')
plt.ylabel('No. of People Who Use')
plt.tight_layout()
# plt.xkcd() #comic style

#bar plot by default direction (vertical)
plt.bar(languages,people)
```

```python
#prepare the plot parameter
plt.title('The Most Popular Languages')
# plt.ylabel('Programing Language')
plt.xlabel('No. of People Who Use')
plt.tight_layout()

#bar plot by horizontal direction and reverse
languages.reverse()
people.reverse()
plt.barh(languages,people)
```
